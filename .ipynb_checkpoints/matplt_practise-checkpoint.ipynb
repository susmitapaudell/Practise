```python
import matplotlib.pyplot as plt
days = range(1,9)
celcius_values = [22,19,43,33,65,77,45,65]

fig, ax = plt.subplots()
ax.plot(days, celcius_values)
ax.set(title = "temperature",
      xlabel = "days",
      ylabel = "temperature on celcius")
plt.show()
```

```python
import matplotlib.pyplot as plt

days = range(5,15)
t_min = [5, 5, 5, 5, 5, 6, 5, 5, 5, 5]
t_max = [19, 19, 20, 20, 19, 19, 18, 18, 18, 15]

fig, ax = plt.subplots()
ax.plot(days, t_min,
        days, t_min, "or",
        days, t_max,
        days, t_max, "oy")
ax.set(title = "temperature of kathmandu from 4th to 14th december, 2024",
      xlabel = "date",
      ylabel = "temperature in celcius")
```
